# ウォレットとは
ウォレットは暗号資産そのものを保管しているわけではなく、その資産に紐づく秘密鍵を保管している。
## ウォレットは秘密鍵を管理している。
公開鍵やアドレスは秘密鍵から生成されるため、秘密鍵は非常に重要なデータであり、秘密鍵を持っていることは資産を持っていることとほぼ同義とも言える。  
ウォレットには形態によってさまざまな種類がある。  
例えば、ウェブ上で利用できる利便性の高いウェブウォレット、PCなどで利用できるデスクトップウォレット、USBのような端末で利用できるハードウェアウォレット、紙に印字して利用するペーパーウォレットなど。  
さらに、ウォレットの状態によって、インターネットに接続しているホットウォレットとオフライン状態で管理されるコールドウォレットにも大きく分類できる。  
## ウォレットの安全性と利便性
ウォレットには重要なデータが収められているため、その安全性には自ずと注目が集まる。また、送金したり受け取ったりするような資産の移動も頻繁に起こるため、利便性も重要視される。  
安全でかつ利便性も高いウォレットが一番ありがたいが、基本的にこれら２つが両立することはない。  
例えば、ホットウォレットの場合、常にインターネットに接続しているため、送金などの資産の移動が楽に行える。しかし、インターネットを経由する攻撃を受けやすく、安全性が比較的低くなってしまう傾向がある。  
実際に、仮想通貨取引所から暗号資産が流出した事件ではホットウォレットをハッキングした攻撃がほとんどであった。  
一方、コールドウォレットはオフラインで管理しているため、外部からハッキングされる心配がほぼない。しかし、送金などの資産の移動をするにはオンラインに一旦戻さなければならないので利便性が高くない。  
このように、秘密鍵を保管するという機能は同じであるものの、その形態によって安全性や利便性に特徴が生まれます。自分の使い方にあったウォレットを選択する必要がある。  
## 秘密鍵をより扱いやすくするために
上記の通り、ウォレットは秘密鍵を管理するものだ。秘密鍵を管理する形態やインターネットへの接続の有無によって、さまざまな種類のウォレットが開発され、それぞれに長所短所があるので、状況によって使い分ける必要がある。  
しかし、どのウォレットを使ったとしても、安全性と利便性のバランスを考えながら秘密鍵を扱わなければならない煩雑さは変わらない。例えば、秘密鍵はコールドウォレット、公開鍵やアドレスはホットウォレットとそれぞれ管理できれば、安全性や利便性を共に高めることができるようになる。そこで、決定性ウォレットやHDウォレットなどさまざまな方式が考案された。

# 非決定性ウォレットと決定性ウォレット
## 非決定性ウォレット
非決定性ウォレットとは、ウォレットないで秘密鍵から公開鍵を１対１の関係で生成する仕組みのウォレット。  
公開鍵の数だけ秘密鍵が必要となり、ウォレットが管理すべき秘密鍵の数が多くなる。ランダムに公開鍵を生成するため、ランダムウォレットと呼ぶこともある。  
非決定性ウォレットの場合、鍵同士に関係性がないため、秘密鍵を紛失した際、それに関連する公開鍵やビットコインん自体を誰も使えなくなる。  
また、公開鍵と秘密鍵が１対１でリンクしているので、一度紛失すると秘密鍵を復旧するのは極めて困難。  
## 決定性ウォレット
決定性ウォレットとは、「`シード`」と呼ばれる１つの乱数を元に、複数の鍵を生成する仕組みのウォレット。親に当たる秘密鍵から子供に当たる秘密鍵を次々に生成していくことで鍵同士に依存関係が生まれる。  
依存関係があるので大元のシードさえバックアップを取っておけば、全ての秘密鍵を復元できる。  
決定性ウォレットは非決定性ウォレットと違い、シードのバックアップがしっかり撮れていれば鍵を復元できるため、ウォレット自体のエクスポートやインポートが容易にできるようになる。  
これはウォレットの利便性を高めれれるとして注目された。そして決定性ウォレットをさらに技術的に推し進めたのが階層的決定性ウォレット(HDウォレット)。

# 階層的決定性ウォレット(HDウォレット)
階層的決定性ウォレットを使うと、秘密鍵や公開鍵の管理コストがさらに下がるので使い勝手がより良くなる！
## HD(解像的決定性)ウォレットの概要
HDウォレットは決定性ウォレットと同じように、シードと呼ばれる乱数から多くの鍵を生成する。決定性ウォレットとの違いは１つの親鍵から生成される子鍵が複数である点。  
これは、鍵群の関係性がツリー構造になるため、ブランチを特定の用途にのみ利用するというように目的に応じて、鍵を使い分けることもできる。  
また、親公開鍵から子秘密鍵を生成できるため、秘密鍵にいたずらにアクセスする必要がなく、セキュリティが高まるというメリットもある。  
## マスター秘密鍵、マスター公開鍵、チェーンコード
HDウォレットにおいて最初に生成される秘密鍵、公開鍵をそれぞれマスター秘密鍵、マスター公開鍵と呼ぶ。  
データとキーという２つのデータを HMAC-SHA512 という関数に入力し、その出力のハッシュ値を前半と後半に256ビットずつに分けたときの前半部分がマスター秘密鍵、後半部分がマスターチェーンコードとなる。  
チェーンコードとは子鍵生成する際に、HMAC-SHA512の2つの出力のうち、キーとして入力されるデータのこと。
## マスター鍵を生成してみる。


```python
import os
import binascii
import ecdsa
import hmac
import hashlib

# シード(データ)の32バイトの乱数を生成
seed = os.urandom(32)
# キーを決める。(ビットコインでは "Bitcoin seed" で固定されている。)
root_key = b"Bitcoin seed"

# HMAC-SHA512を実行する関数
def hmac_sha512(data, keymessage):
    hash = hmac.new(data,keymessage, hashlib.sha512).digest()
    return hash

# 秘密鍵から楕円曲線暗号を通して公開鍵を生成する関数
def create_pubkey(private_key):
    publickey = ecdsa.SigningKey.from_string(private_key, curve=ecdsa.SECP256k1).verifying_key.to_string()
    return publickey

# seedをデータ、root_keyをキーとして HMAC-SHA512 を実行。
master = hmac_sha512(seed, root_key)
# 出力された512ビット(64バイト)のうち、
# 前半部分をマスター秘密鍵
master_secratkey = master[:32]
# 後半部分をマスターチェーンコード
master_chaincode = master[32:]
# 公開鍵の生成
master_publickey = create_pubkey(master_secratkey)
# y座標を取り出す。
master_publickey_integer = int.from_bytes(master_publickey[32:], byteorder="big")

# 圧縮公開鍵生成
if master_publickey_integer % 2 == 0:
    master_publickey_x = b"\x02" + master_publickey[:32]
else:
    master_publickey_x = b"\x03" + master_publickey[:32]

# マスター秘密鍵
print("マスター秘密鍵")
print(binascii.hexlify(master_secratkey))

# マスターチェーンコード
print("マスターチェーンコード")
print(binascii.hexlify(master_chaincode))

# マスター圧縮公開鍵
print("マスター公開鍵")
print(binascii.hexlify(master_publickey_x))
```

## 子鍵の生成
生成されたチェーンコードは子鍵を生成する際にキーとしてHMAC-SHA512に入力される。  
子鍵(子秘密鍵と子公開鍵)を生成するプロセスはデータの生成過程と子秘密鍵の生成過程がマスターキーやマスターチェーンコードを生成する過程と異なっている。  
まず、HMAC-SHA512 にインプットされるデータとして、親の公開鍵の先頭にインデックスを結合したものを利用する。  
これによって生成される鍵が何番目の鍵かを表すことができる。また、子秘密鍵を生成する際には、親秘密鍵とインデックス、HMAC-SHA512の出力結果の前半256ビット(32バイト)を全て足し合わせることで計算する。  
なお、HMAC-SHA512の後半は「子鍵の子鍵(孫鍵)」を生成する際に、HMAC-SHA512のキーとして利用される。  
このように、連鎖的に鍵が生成されることで、相互に依存関係を持つ鍵を大量に効率的に生成できるようになる。  
## 子秘密鍵の生成をしてみる

```python
#####  さっきの続き ######

index = 0
index_bytes = index.to_bytes(8, "big")
# 生成済みの公開鍵とインデックスを結合
data = master_publickey_x + index_bytes
# HMAC-SHA512にデータとチェーンコードをインプットし、出力のハッシュ値を取得
result_hmac512 = hmac_sha512(data, master_chaincode)

# 親秘密鍵をHMAC-SHA512の結果の前半部分を足し合わせる(ハッシュ値はバイト型になっているので整数型に変換。)
sum_integer = int.from_bytes(master_secratkey, "big") + int.from_bytes(result_hmac512[:32], "big")

#　秘密鍵が32バイトよりも大きくならないように巨大な素数であるpで割ってその余りを子秘密鍵として変数へ格納。(詳細は高度な数学なので他で勉強しろ)
p = 2**256 - 2**32 - 2**9 - 2**8 - 2**7 - 2**6 - 2**4 - 1
child_secratkey = (sum_integer % p).to_bytes(32, "big")

# 子秘密鍵(マスターから見て１つ下の階層の秘密鍵)
print("子秘密鍵")
print(binascii.hexlify(child_secratkey))
```

## 拡張鍵
拡張鍵は秘密鍵や公開鍵にチェーンコードをはじめとした情報をセットにした鍵で、通常の秘密鍵や公開鍵よりも情報面でも機能面でも拡張されているためこのように名付けられた。  
拡張鍵はHDウォレットを異なるサーバー噛んで移動可能にするために導入された。  
HDウォレットの子鍵を生成するには親鍵とチェーンコード、インデックスなどが必要で、親鍵だけでは子鍵を生成することができない。 
そのため、親鍵の情報と子鍵の生成に必要なチェーンコードなどのデータを一まとめにした鍵を作成することで移植を容易にし、その鍵以降の子世代の鍵群を生成することができる。  
　　

拡張鍵のフォーマット
- Version bytes: 
- Depth: マスターを0x00とした時の深さ
- Fingerprint: 親公開鍵のHASH160の先頭4バイト
- Chain code: チェーンコード
- Key: 圧縮公開鍵か秘密鍵
- Checksum: ミスや不正を検出するためのチェックサム

Keyの部分に圧縮公開鍵を入れた場合は`拡張公開鍵`、秘密鍵を入れた場合は`拡張秘密鍵`となる。  
拡張公開鍵からは子公開鍵が生成でき、秘密鍵が必要ないので、セキュリティを維持しつつアドレスを生成することが可能。  
このような性質を利用することで、コールドウォレットとホットウォレットで鍵を分散させて管理ができる。  
シードやマスター秘密鍵などの重要な情報はコールドウォレットで管理し、小口の取引用や高額取引用の公開鍵はホットウォレットで管理するといったことが可能。

## 強化導出鍵
`強化導出鍵`は子鍵を生成する際に、HMAC-SHA512のデータとして秘密鍵を利用するもの。  
拡張鍵で圧縮公開鍵を利用するときには、HMAC-SHA512のデータとして公開鍵を利用していたが、この方式は拡張鍵を利用して第三者のサービス上で子公開鍵やアドレスを生成している状況の時に、親公開鍵と親チェーンコードが知られてしまうことになるので、子秘密鍵が何らかの理由で流出してしまうと、同じ階層の全ての秘密鍵が特定されてしまうというリスクがある。  
これは子秘密鍵が親秘密鍵とその他の情報が単に足し合わされただけのため、すでに知られている親公開鍵と親チェーンコードなどを利用すると、  
子秘密鍵 = 親秘密鍵 + その他の情報 → 親秘密鍵　= 子秘密鍵 - その他の情報  
というように逆算すると、親秘密鍵が特定できてしまうため。  
これでは、たとえ親秘密鍵をコールドウォレットに保管していたとしても、簡単に第三者に知られてしまう。  
このセキュリティ面への対策のために、公開鍵ではなく秘密鍵をデータとして利用することが提案され、それを強化導出鍵と呼ぶ。
## HDウォレットへのパス
HDウォレットでは、URLのようにパスを用いてツリー構造における鍵の所在を特定する。  
可読性を高めるためにスラッシュを使って世代ごとに区切ることで構造を表現する。  この時のマスター秘密鍵は m, マスター公開鍵は M, で表現される。  
また、通常の鍵と強化導出鍵はアポストロフィの有無で区別し、通常の鍵は 0, 強化導出鍵は 0', と表現する。  
例えば、 m/0 はマスター秘密鍵から生成された最初の子秘密鍵、M/2 は３番目のこの公開鍵、 m/0/1' は最初の子の２番目の強化導出された子の秘密鍵　を示す。  
しかし、無尽蔵に階層が深くなっていくと管理が大変なので、各階層に意味を持たせることで管理しやすくするアイデアがある。  
例えば、 m/purpose'/coin_type'/account'/change/address_index   
purpose' でウォレットの目的を表し、44か49が設定されている。44は階層の意味をあらかじめ定義するという提案がBIP44で行われたことに由来している。49はSegWitが導入されたことにより設定された。  
coint_type' は仮想通貨を特定するための情報が設定される。ビットコインのメインネットの場合は0, ビットコインのテストネットの場合は1, ライトコインでは2, イーサリアムでは60, が設定される。各仮想通貨への数字の割り当てはBIP44によって事前に定義されており、新しいコインをcoin_typeに登録することも可能で、新規にコインへインデックスが割り当てられる。  
account' は口座に該当し、個人用、組織用といった形で使い分けることが可能。  
change では０と１のみが設定でき、それぞれ受け取りようとお釣り用の２種類が設定できる。  
address_index　の層の公開鍵で実際のアドレスを生成する。  

BIP44に基づくパス表現の例
M/44'/0'/0'/0/0: 最初のビットコイン口座に対する最初の受け取り用の公開鍵  
M/44'/0'/1'/1/1: ２番目のビットコイン口座に対する２番目のお釣り用の公開鍵  
M/44'/60'1'/1/19: ２番目のイーサリアム口座に対する２０番目のおつり用口座の公開鍵  

BIP44: Bitcoin Improvement Proposalsの略でビットコインをより良くするために世界中のエンジニアや研究者から提案される草案のこと。  
SegWit: 取引データから署名部分をwithnessと呼ばれる別の領域に格納することで、署名の正しさを変更することなく取引データを改竄できないようにしたり、取引データのサイズを圧縮したりする技術のこと。  


```python
import os
import binascii
import ecdsa
import hmac
import hashlib

# シード(データ)の32バイトの乱数を生成
seed = os.urandom(32)
# キーを決める。(ビットコインでは "Bitcoin seed" で固定されている。)
root_key = b"Bitcoin seed"

# HMAC-SHA512を実行する関数
def hmac_sha512(data, keymessage):
    hash = hmac.new(data,keymessage, hashlib.sha512).digest()
    return hash

# 秘密鍵から楕円曲線暗号を通して公開鍵を生成する関数
def create_pubkey(private_key):
    publickey = ecdsa.SigningKey.from_string(private_key, curve=ecdsa.SECP256k1).verifying_key.to_string()
    return publickey

# seedをデータ、root_keyをキーとして HMAC-SHA512 を実行。
master = hmac_sha512(seed, root_key)
# 出力された512ビット(64バイト)のうち、
# 前半部分をマスター秘密鍵
master_secratkey = master[:32]
# 後半部分をマスターチェーンコード
master_chaincode = master[32:]
# 公開鍵の生成
master_publickey = create_pubkey(master_secratkey)
# y座標を取り出す。
master_publickey_integer = int.from_bytes(master_publickey[32:], byteorder="big")

# 圧縮公開鍵生成
if master_publickey_integer % 2 == 0:
    master_publickey_x = b"\x02" + master_publickey[:32]
else:
    master_publickey_x = b"\x03" + master_publickey[:32]

# マスター秘密鍵
print("マスター秘密鍵")
print(binascii.hexlify(master_secratkey))

# マスターチェーンコード
print("マスターチェーンコード")
print(binascii.hexlify(master_chaincode))

# マスター圧縮公開鍵
print("マスター公開鍵")
print(binascii.hexlify(master_publickey_x))

index = 1
index_bytes = index.to_bytes(8, "big")
# 生成済みの公開鍵とインデックスを結合
data = master_publickey_x + index_bytes
# HMAC-SHA512にデータとチェーンコードをインプットし、出力のハッシュ値を取得
result_hmac512 = hmac_sha512(data, master_chaincode)

# 親秘密鍵をHMAC-SHA512の結果の前半部分を足し合わせる(ハッシュ値はバイト型になっているので整数型に変換。)
sum_integer = int.from_bytes(master_secratkey, "big") + int.from_bytes(result_hmac512[:32], "big")

#　秘密鍵が32バイトよりも大きくならないように巨大な素数であるpで割ってその余りを子秘密鍵として変数へ格納。(詳細は高度な数学なので他で勉強しろ)
p = 2**256 - 2**32 - 2**9 - 2**8 - 2**7 - 2**6 - 2**4 - 1
child_secratkey = (sum_integer % p).to_bytes(32, "big")

# 子秘密鍵(マスターから見て１つ下の階層の秘密鍵)
print("\n")
print("子秘密鍵")
print(binascii.hexlify(child_secratkey))


child_chaincode = result_hmac512[32:]
child_publickey = create_pubkey(child_secratkey)
child_publickey_integer = int.from_bytes(child_publickey[32:], byteorder="big")

# 圧縮子公開鍵生成
if child_publickey_integer % 2 == 0:
    child_publickey_x = b"\x02" + child_publickey[:32]
else:
    child_publickey_x = b"\x03" + child_publickey[:32]


print("子チェーンコード")
print(binascii.hexlify(child_chaincode))
print("子公開鍵")
print(binascii.hexlify(child_publickey_x))
```
